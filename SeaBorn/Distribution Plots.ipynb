```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde = False)
```

```python
sns.distplot(tips['total_bill'], kde = False, bins = 30)
```

```python
sns.jointplot(x ='total_bill', y='tip', data = tips)
```

```python
sns.jointplot(x ='total_bill', y='tip', data = tips, kind = 'hex')
```

```python
sns.jointplot(x ='total_bill', y='tip', data = tips, kind = 'reg')
```

```python
sns.jointplot(x ='total_bill', y='tip', data = tips, kind = 'kde')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'sex') # here we can identify by categories using hue
```

```python
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm') #pallete is used to select different type of color patterns
```

```python
sns.rugplot(tips['total_bill'], height = 0.25)
```

```python
sns.distplot(tips['total_bill']) #KDE = Kernel density estimation
```
