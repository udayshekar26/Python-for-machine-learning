```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris)
```

```python
h = sns.PairGrid(iris)
h.map(plt.scatter)
```

```python
h.map_diag(sns.barplot)
h.map_lower(plt.scatter)
h.map_upper(sns.regplot)
```

```python
g = sns.FacetGrid(data = tips, col='time', row='smoker')
```
