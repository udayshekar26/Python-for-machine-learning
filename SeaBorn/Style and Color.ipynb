```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(x= 'sex', data = tips)
```

```python
sns.set_style('ticks')
sns.countplot(x= 'sex', data = tips)
```

```python
sns.set_style('ticks')
sns.countplot(x= 'sex', data = tips)
sns.despine()
```

```python
sns.set_style('whitegrid')
sns.countplot(x= 'sex', data = tips)
```

```python
plt.figure(figsize =(12,5))
sns.countplot(x= 'sex', data = tips)
```

```python
sns.set_context(context='notebook')
sns.countplot(x='sex',data = tips)
```

```python
sns.lmplot(x='total_bill',y='tip', data = tips, hue= 'sex', palette = 'Spectral')

#different palete link of matplotlib documentation: https://matplotlib.org/2.0.2/examples/color/colormaps_reference.html
```
